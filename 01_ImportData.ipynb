```python


def import_all_csvfiles_into_df(path, drop_dups = False , check_for_word = 'None', DEBUG = False):
    import glob, re
    #path = "D:\OneDrive\Trading\Tweets" # use your path
    all_files = glob.glob(path + "/*.csv")

    li = []

    for filename in all_files:
        if DEBUG : print(f'working through {filename}')
        if check_for_word == 'None':
            print(filename)
            df = pd.read_csv(filename, index_col=None, header=0)
            li.append(df)
        else:
            if bool(re.search(check_for_word, filename)):
                print(f'importing:\t{filename}')
                df = pd.read_csv(filename, index_col=None, header=0)
                li.append(df)
            else:
                pass
                

    alldata = pd.concat(li, axis=0, ignore_index=True)
    if DEBUG : print('Before deduping the shape is ', alldata.shape)
    if drop_dups:
        alldata = pd.concat(li, axis=0, ignore_index=True).drop_duplicates()
        if DEBUG : print('After dedup the shape is ', alldata.shape)
        
    return alldata
```
