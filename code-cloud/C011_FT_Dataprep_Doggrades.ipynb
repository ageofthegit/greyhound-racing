```python

dogs_files = ['ft_dogs_ff_2021_01_01_to_2022_12_14.csv',
                 'ft_dogs_ff_2022_12_15_to_2023_01_30.csv',
                 'ft_dogs_ff_2023_01_31_to_2023_05_07.csv',
                 'ft_dogs_ff_2023_01_31_to_2023_05_07_missing.csv']

#filename_ff_dogs_a = 'ft_dogs_ff_2021_01_01_to_2022_12_14.csv'
#filename_ff_dogs_b = 'ft_dogs_ff_2022_12_15_to_2023_01_30.csv'

#print(f'Importing : {fileloc}/{filename_ff_dogs_a}')
#print(f'Importing : {fileloc}/{filename_ff_dogs_b}')

df_ft_dogs_master = pd.DataFrame()

for dogs in dogs_files:
    print(datetime.now())
    print(f'Importing : {str(fileloc + dogs)}')

    df_ft_dogs_temp = pd.read_csv(f'{str(fileloc + dogs)}')
    print(df_ft_dogs_temp.shape)
    print(df_ft_dogs_temp.columns.values.tolist())
    
    if df_ft_dogs_master.empty:
        df_ft_dogs_master = df_ft_dogs_temp.copy()
    else:
        df_ft_dogs_master = df_ft_dogs_master.append(df_ft_dogs_temp)
        print(df_ft_dogs_master.shape)
    print()
```
