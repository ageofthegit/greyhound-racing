### B) Create Variable for Speed Profile

```python

# Sort for Easy QC
#df_base_sp.sort_values(by = ['@id_DOG', 'DogName', 'Event_Dt'], inplace = True)
df_base_sp.sort_values(by = ['TrainerId', '@id_RACE', 'Event_Dt'], inplace = True)
```

```python

df_base_sp[df_base_sp['TrainerId'].isin([92892])].head()
```

```python

# ADDITIONAL VARIABLES
vtrnr = ['TrainerId']
vrace = ['@id_RACE']
vevnt = ['Event_Dt']

# Distance and Speed in each race
#df_base_sp['x_dist'] = df_base_sp['Distance'].apply(lambda x : x.replace('m','')).astype(int)
df_base_sp['x_dist'] = df_base_sp['Distance'].str.replace('m','').astype(int)
df_base_sp['speed'] = (df_base_sp['x_dist'] / df_base_sp['y_runtime'])*1000

# Flag for getting a total race count
df_base_sp['f_1'] = 1

# Total Races
df_base_sp['x_dograced_tot'] = df_base_sp.groupby(vtrnr).cumcount()
```

```python

#df_base_sp.loc[:,"trnr_wins"] = df_base_sp['y_pos'].apply(lambda x : 1 if x == '1' else 0 )
df_base_sp.loc[:,"trnr_wins"] = df_base_sp['y_pos'].eq('1').astype(int)


#df_base_sp['x_dograced_tot']
```

```python

#df_base_sp.loc[:,"trnr_fin_plc"] = df_base_sp.apply(lambda x : 1 if (x.position in [1,2,3]) & (x.x_num_dogs >=8) \
#else 1 if (x.position in [1,2]) & (x.x_num_dogs >= 5) \
#else 0 if (x.x_num_dogs < 5) \
#else 0, axis = 1)

conditions = [
    ((df_base_sp['position'].isin([1,2,3])) & (df_base_sp['x_num_dogs'] >= 8)), 
    ((df_base_sp['position'].isin([1,2])) & (df_base_sp['x_num_dogs'] >= 5)), 
    (df_base_sp['x_num_dogs'] < 5)
]

choices = [1, 1, 0]

df_base_sp['trnr_fin_plc'] = np.select(conditions, choices, default=0)
```

```python

#df_base_sp.loc[:,"trnr_fin_t2"] = df_base_sp['position'].apply(lambda x : 1 if x in [1,2] else 0 )
df_base_sp['trnr_fin_t2'] = df_base_sp['position'].isin([1, 2]).astype(int)
```

```python

#df_base_sp.loc[:,"trnr_fin_last"] = df_base_sp.apply(lambda x : 1 if x.y_pos == x.x_num_dogs else 0 , axis = 1)
df_base_sp["trnr_fin_last"] = (df_base_sp['y_pos'] == df_base_sp['x_num_dogs']).astype(int)
```

```python

#df_base_sp.loc[:,"trnr_fin_last2"] = df_base_sp.apply(lambda x : 1 if (x.position == x.x_num_dogs) | (x.position == x.x_num_dogs - 1) else 0 , axis = 1)
df_base_sp['trnr_fin_last2'] = ((df_base_sp['position'] == df_base_sp['x_num_dogs']) | 
                                (df_base_sp['position'] == df_base_sp['x_num_dogs'] - 1)).astype(int)
```

```python

print(df_base_sp.columns.values.tolist())
```

```python

agg_dict = {
    '@id_DOG': 'nunique',
    'trnr_wins': 'sum',
    'trnr_fin_plc': 'sum',
    'trnr_fin_t2': 'sum',
    'trnr_fin_last': 'sum',
    'trnr_fin_last2': 'sum'
}

#df_base_trnr = df_base_sp.groupby(vtrnr + vrace + vevnt).agg( {'@id_DOG':'nunique', 'trnr_wins':'sum', 'trnr_fin_plc':'sum', 'trnr_fin_t2':'sum', 'trnr_fin_last':'sum', 'trnr_fin_last2':'sum'} ).reset_index().rename(columns = {'@id_DOG':'trnr_num_dogs'})
df_base_trnr = df_base_sp.groupby(vtrnr + vrace + vevnt).agg(agg_dict).reset_index().rename(columns={'@id_DOG': 'trnr_num_dogs'})

print(type(df_base_trnr))
print(df_base_trnr.columns.values.tolist())
```

```python

qc_rowcounts = df_base_trnr.groupby('trnr_num_dogs', dropna = False).agg({'@id_RACE':'count'}).reset_index().rename(columns = {'@id_RACE':'trnr_races'})
```

```python

df_base_trnr.sort_values(by = ['TrainerId','Event_Dt','@id_RACE'], inplace = True)

print(datetime.now())
```

```python

df_base_trnr['trnr_xr_races_tot'] = df_base_trnr.groupby(vtrnr).cumcount()
```

```python

df_base_trnr['trnr_xr_dogsraced_tot'] = df_base_trnr.groupby(vtrnr)['trnr_num_dogs'].cumsum()
```

```python

df_base_trnr[df_base_trnr.TrainerId.isin([6418])].tail(25)
```

```python

df_base_trnr.loc[:,"trnr_xr_races_tot_prev"] = df_base_trnr.groupby(vtrnr)['trnr_xr_races_tot'].shift(1)
df_base_trnr["trnr_xr_races_tot_prev"].fillna(1, inplace = True)

df_base_trnr.loc[:,"trnr_xr_dogsraced_tot_prev"] = df_base_trnr.groupby(vtrnr)['trnr_xr_dogsraced_tot'].shift(1)
df_base_trnr["trnr_xr_dogsraced_tot_prev"].fillna(1, inplace = True)


df_base_trnr.loc[:,"trnr_wins_prev"] = df_base_trnr.groupby(vtrnr)['trnr_wins'].shift(1)
df_base_trnr["trnr_wins_prev"].fillna(0, inplace = True)

df_base_trnr.loc[:,"trnr_fin_plc_prev"] = df_base_trnr.groupby(vtrnr)['trnr_fin_plc'].shift(1)
df_base_trnr["trnr_fin_plc_prev"].fillna(0, inplace = True)
df_base_trnr.loc[:,"trnr_fin_t2_prev"] = df_base_trnr.groupby(vtrnr)['trnr_fin_t2'].shift(1)
df_base_trnr["trnr_fin_t2_prev"].fillna(0, inplace = True)

df_base_trnr.loc[:,"trnr_fin_last_prev"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last'].shift(1)
df_base_trnr["trnr_fin_last_prev"].fillna(0, inplace = True)
df_base_trnr.loc[:,"trnr_fin_last2_prev"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last2'].shift(1)
df_base_trnr["trnr_fin_last2_prev"].fillna(0, inplace = True)

print(datetime.now())
```

```python

df_base_trnr.loc[:,"trnr_xr_wins"] = df_base_trnr.groupby(vtrnr)['trnr_wins_prev'].rolling(window=999, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xr_fin_plc"] = df_base_trnr.groupby(vtrnr)['trnr_fin_plc_prev'].rolling(window=999, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xr_fin_t2"] = df_base_trnr.groupby(vtrnr)['trnr_fin_t2_prev'].rolling(window=999, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xr_fin_last"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last_prev'].rolling(window=999, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xr_fin_last2"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last2_prev'].rolling(window=999, min_periods = 1).sum().reset_index(0,drop=True)

df_base_trnr['trnr_xr_wins'].fillna(0, inplace = True)
df_base_trnr['trnr_xr_fin_plc'].fillna(0, inplace = True)
df_base_trnr['trnr_xr_fin_t2'].fillna(0, inplace = True)
df_base_trnr['trnr_xr_fin_last'].fillna(0, inplace = True)
df_base_trnr['trnr_xr_fin_last2'].fillna(0, inplace = True)

print(datetime.now())
```

```python

df_base_trnr.loc[:,"trnr_xrl5_wins"] = df_base_trnr.groupby(vtrnr)['trnr_wins_prev'].rolling(window=5, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xrl5_fin_plc"] = df_base_trnr.groupby(vtrnr)['trnr_fin_plc_prev'].rolling(window=5, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xrl5_fin_t2"] = df_base_trnr.groupby(vtrnr)['trnr_fin_t2_prev'].rolling(window=5, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xrl5_fin_last"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last_prev'].rolling(window=5, min_periods = 1).sum().reset_index(0,drop=True)
df_base_trnr.loc[:,"trnr_xrl5_fin_last2"] = df_base_trnr.groupby(vtrnr)['trnr_fin_last2_prev'].rolling(window=5, min_periods = 1).sum().reset_index(0,drop=True)

df_base_trnr['trnr_xrl5_wins'].fillna(0, inplace = True)
df_base_trnr['trnr_xrl5_fin_plc'].fillna(0, inplace = True)
df_base_trnr['trnr_xrl5_fin_t2'].fillna(0, inplace = True)
df_base_trnr['trnr_xrl5_fin_last'].fillna(0, inplace = True)
df_base_trnr['trnr_xrl5_fin_last2'].fillna(0, inplace = True)

print(datetime.now())
```

```python

# Divide by Total Races
df_base_trnr.loc[:,"trnr_xr_wins_pct"] = df_base_trnr['trnr_xr_wins'] / df_base_trnr['trnr_xr_races_tot_prev']
df_base_trnr['trnr_xr_wins_pct'].fillna(0, inplace = True)

# Divide by Total Dogs
df_base_trnr.loc[:,"trnr_xr_fin_t2_pct"] = df_base_trnr['trnr_xr_fin_t2'] / df_base_trnr['trnr_xr_dogsraced_tot_prev']
df_base_trnr['trnr_xr_fin_t2_pct'].fillna(0, inplace = True)

df_base_trnr.loc[:,"trnr_xr_fin_plc_pct"] = df_base_trnr['trnr_xr_fin_plc'] / df_base_trnr['trnr_xr_dogsraced_tot_prev']
df_base_trnr['trnr_xr_fin_plc_pct'].fillna(0, inplace = True)


# Divide by Total Races
df_base_trnr.loc[:,"trnr_xr_fin_last_pct"] = df_base_trnr['trnr_xr_fin_last'] / df_base_trnr['trnr_xr_races_tot_prev']
df_base_trnr['trnr_xr_fin_last_pct'].fillna(0, inplace = True)

# Divide by Total Dogs
df_base_trnr.loc[:,"trnr_xr_fin_last2_pct"] = df_base_trnr['trnr_xr_fin_last2'] / df_base_trnr['trnr_xr_dogsraced_tot_prev']
df_base_trnr['trnr_xr_fin_last2_pct'].fillna(0, inplace = True)

print(datetime.now())
```
