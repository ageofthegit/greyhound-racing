```python

# ADDITIONAL VARIABLES
gpbylevel = ['@id_DOG']

# Distance and Speed in each race
df.ft['x_dist'] = df.ft['Distance'].apply(lambda x : x.replace('m','')).astype(int)
df.ft['speed'] = (df.ft['x_dist'] / df.ft['RunTime'])*1000

# Average Speed until prev race
## CAN USE THIS CODE TO GET AVERAGE SPEED L 1/3/5/7/10 races
df.ft['df_avg_speed_prev'] = df.ft.groupby(gpbylevel)['speed'].expanding().mean().shift(1).reset_index().set_index('level_1').rename(columns = {'speed':'avg_speed_prev'})['avg_speed_prev']
# 2 Min Runtime

# Flag for getting a total race count
df.ft['f_1'] = 1

# Total Races
df.ft['x_races_tot'] = df.ft.groupby(gpbylevel).cumcount()
```

```python
# Fix the Place Variable

#print(df.ft.Place.value_counts())

df.ft['position'] = df.ft.Place.apply(lambda x : x.replace("=","") if x in ['1=','2=','3=','4=','5=','6=','7=','8='] else x)

#print(df.ft.position.value_counts())
```
